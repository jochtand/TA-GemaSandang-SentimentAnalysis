```python
import pandas as pd
import random

print("Generating Fashion Trend Dataset...")

# =====================================================
# Bagian umum
# =====================================================

awal = [
    "awalnya",
    "pas pertama lihat",
    "jujur",
    "kalau menurut saya",
    "sejujurnya",
    "entah kenapa",
    "buat saya pribadi",
    "kesan pertama",
    "waktu pertama lihat",
    "setelah beberapa kali lihat"
]

objek = [
    "modelnya",
    "warnanya",
    "stylenya",
    "outfitnya",
    "looknya",
    "konsepnya",
    "desainnya",
    "kombinasi warnanya"
]

kata_bagus = [
    "bagus",
    "menarik",
    "keren",
    "unik",
    "oke",
    "cukup bagus",
    "lumayan",
    "estetik"
]

kata_kurang = [
    "kurang menarik",
    "kurang cocok",
    "terlalu biasa",
    "agak monoton",
    "kurang estetik",
    "membosankan"
]

penutup_random = [
    "",
    "",
    "",
    "sih",
    "jujur",
    "imo",
    "haha",
    "wkwk",
    "menurutku"
]

# =====================================================
# Positif
# =====================================================

positif_akhir = [
    "jadi suka",
    "makin menarik",
    "ternyata keren",
    "ternyata cocok",
    "malah estetik",
    "jadi pengen coba",
    "ternyata bagus juga",
    "makin dilihat makin menarik"
]

positif_personal = [
    "biasanya saya kurang suka model begini tapi yang ini menarik",
    "bukan style yang biasa saya pakai tapi saya suka",
    "entah kenapa jadi tertarik",
    "saya pribadi suka konsepnya",
    "makin dilihat makin cocok",
    "ga nyangka ternyata bagus"
]

# =====================================================
# Negatif
# =====================================================

negatif_akhir = [
    "jadi biasa aja",
    "kurang cocok",
    "makin membosankan",
    "ga sesuai ekspektasi",
    "ternyata kurang menarik",
    "terasa monoton",
    "kurang masuk di saya"
]

negatif_personal = [
    "banyak yang suka tapi saya kurang tertarik",
    "mungkin cocok buat orang lain tapi bukan saya",
    "saya paham kenapa banyak yang suka tapi kurang cocok buat saya",
    "buat saya pribadi kurang menarik",
    "saya masih belum suka model seperti ini"
]

# =====================================================
# Netral
# =====================================================

netral_akhir = [
    "tergantung selera",
    "belum tentu saya pakai",
    "ga ada yang spesial",
    "masih standar",
    "lumayan aja",
    "biasa aja menurut saya",
    "ga buruk juga"
]

netral_personal = [
    "mungkin cocok buat sebagian orang",
    "sulit bilang bagus atau tidak",
    "saya masih netral melihatnya",
    "belum punya pendapat khusus",
    "tidak jelek tapi juga tidak spesial",
    "masih biasa aja menurut saya"
]

# =====================================================
# Slang
# =====================================================

kamus_slang = {
    "bagus": ["bgs", "baguss"],
    "keren": ["krn", "kereen"],
    "menarik": ["menarikk"],
    "saya": ["aku", "gue", "gw"],
    "tidak": ["ga", "gak"],
    "cukup": ["cukup", "lumayan"],
}

def augmentasi(teks):

    hasil = []

    for kata in teks.split():

        kata_bersih = kata.lower().replace(",", "")

        if kata_bersih in kamus_slang and random.random() < 0.25:
            hasil.append(random.choice(kamus_slang[kata_bersih]))
        else:
            hasil.append(kata)

    teks = " ".join(hasil)

    tambahan = random.choice(penutup_random)

    if tambahan != "":
        teks += " " + tambahan

    return teks

# =====================================================
# Generator Positif
# =====================================================

def generate_positif():

    pola = random.randint(1, 6)

    if pola == 1:
        return f"{random.choice(awal)} {random.choice(kata_kurang)} tapi {random.choice(positif_akhir)}"

    elif pola == 2:
        return random.choice(positif_personal)

    elif pola == 3:
        return f"{random.choice(objek)} {random.choice(kata_bagus)}"

    elif pola == 4:
        return f"{random.choice(objek)} {random.choice(kata_bagus)} tapi tetap terlihat menarik"

    elif pola == 5:
        return f"{random.choice(awal)} biasa aja tapi {random.choice(positif_akhir)}"

    else:
        return f"awalnya kurang yakin tapi {random.choice(positif_akhir)}"

# =====================================================
# Generator Negatif
# =====================================================

def generate_negatif():

    pola = random.randint(1, 6)

    if pola == 1:
        return f"{random.choice(awal)} {random.choice(kata_bagus)} tapi {random.choice(negatif_akhir)}"

    elif pola == 2:
        return random.choice(negatif_personal)

    elif pola == 3:
        return f"{random.choice(objek)} {random.choice(kata_kurang)}"

    elif pola == 4:
        return f"{random.choice(objek)} {random.choice(kata_bagus)} tapi menurut saya {random.choice(kata_kurang)}"

    elif pola == 5:
        return f"banyak yang bilang keren tapi {random.choice(negatif_akhir)}"

    else:
        return f"awalnya suka tapi {random.choice(negatif_akhir)}"

# =====================================================
# Generator Netral
# =====================================================

def generate_netral():

    pola = random.randint(1, 6)

    if pola == 1:
        return f"{random.choice(objek)} {random.choice(kata_bagus)} tapi {random.choice(netral_akhir)}"

    elif pola == 2:
        return random.choice(netral_personal)

    elif pola == 3:
        return f"{random.choice(objek)} {random.choice(kata_bagus)}"

    elif pola == 4:
        return f"{random.choice(awal)} {random.choice(kata_bagus)} tapi {random.choice(netral_akhir)}"

    elif pola == 5:
        return f"{random.choice(objek)} terlihat oke tapi {random.choice(netral_akhir)}"

    else:
        return f"menarik sih tapi {random.choice(netral_akhir)}"

# =====================================================
# Generate Dataset
# =====================================================

data = []

for _ in range(7000):
    data.append({
        "ulasan": augmentasi(generate_positif()),
        "sentimen": "Positif"
    })

for _ in range(7000):
    data.append({
        "ulasan": augmentasi(generate_negatif()),
        "sentimen": "Negatif"
    })

for _ in range(6000):
    data.append({
        "ulasan": augmentasi(generate_netral()),
        "sentimen": "Netral"
    })

random.shuffle(data)

df = pd.DataFrame(data)

df.to_csv("dataset_trenfashion.csv", index=False)

print(df["sentimen"].value_counts())

print("\nUnique Total:")
print(df["ulasan"].nunique())

print("\nUnique Per Class:")
print(df.groupby("sentimen")["ulasan"].nunique())
```
